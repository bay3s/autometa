```python
import pandas as pd
import seaborn as sns

import numpy as np
from matplotlib import pyplot as plt
```

```python
heatmap_data = pd.read_csv("./heatmap-results.csv")
ax = sns.kdeplot(data=heatmap_data, x="x", y="y", fill=True, thresh=0, levels=10, cmap="mako")
plt.show()
```

```python
plt.style.use('seaborn-white')
plt.contour(heatmap_data["x"], heatmap_data["y"], heatmap_data["r"], levels=10)
plt.show()
```

```python
fig = ax.get_figure()
fig.savefig('kde.png', transparent=True, bbox_inches='tight', pad_inches=0)
```
